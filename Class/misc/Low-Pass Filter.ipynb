```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 20, 200)

y = 2*np.sin(x) + 1 + np.random.randn(200)

plt.figure(figsize = (8,6))
plt.plot(x, y)
```

```python
y_smoothed = [y[0]]
alpha = 0.99999999

for i in range(1,len(x)):
    y_smoothed.append(alpha*y[:i].mean() + (1 - alpha)*y[i])
```

```python
plt.figure(figsize = (8,6))
plt.plot(x, y, label = "Original Signal")
plt.plot(x, y_smoothed, label = "Smoothed Signal")
plt.legend()
```

```python
y_smoothed = [0, y[0]]
alpha = 0.999

for i in range(1,len(x)-1):
    y_smoothed.append(alpha*y[:i].mean() + (1 - alpha)*y[i])
```

```python
plt.figure(figsize = (8,6))
plt.plot(x, y, label = "Original Signal")
plt.plot(x, y_smoothed, label = "Smoothed Signal")
plt.legend()
```
